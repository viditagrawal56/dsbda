```python
##USING MATPLOTLIB
import matplotlib.pyplot as plt

#LINE
plt.plot([1,2,3,4],[1,4,9,16],"g--")
plt.ylabel('some Number')
plt.xlabel('any data')
plt.show()

plt.subplot(221)
#PIE
cars=['AUDI','BMW','FORD','TESLA','JAGUR','MERCEDES']
data=[23,17,35,29,12,41]
fig=plt.figure(figsize=(10,7))
plt.pie(data,labels=cars)
plt.show()

#BAR
plt.bar(cars,data)
plt.show()

#SCATTER
plt.scatter(cars,data)
plt.show()


##USING SEABORN
import seaborn as sns
import matplotlib.pyplot as plt

cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']
data = [23, 17, 35, 29, 12, 41]
import pandas as pd
df = pd.DataFrame({'Cars': cars, 'Data': data})

# Plotting Line Plot
plt.figure(figsize=(10, 5))
sns.lineplot(x='Cars', y='Data', data=df)
plt.title('Line Plot')
plt.xticks(rotation=45)
plt.show()

# Plotting Scatter Plot
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Cars', y='Data', data=df, color='blue')
plt.title('Scatter Plot')
plt.xticks(rotation=45)
plt.show()

# Plotting Bar Plot
plt.figure(figsize=(10, 5))
sns.barplot(x='Cars', y='Data', data=df, color='green')
plt.title('Bar Plot')
plt.xticks(rotation=45)
plt.show()

# Plotting Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(df['Data'], labels=df['Cars'], autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()

```
